```python
# convert xlsx into dataframe, remove first row and headers
df = pd.read_excel("NZ_Admin_JOBS.xlsx", skiprows = 1, header=None)
```

```python
# column 3 replace empty cells with N/A
df.fillna('N/A', inplace=True)
```

```python
# check column 3 info before modification
df[3].head(40)
```

```python
# split location and area infomation into two different columns 
# remove all none infomation related items 
df[['location','area']] = df[3].str.split('area:', expand=True)
df['location'] = df['location'].str.replace('location:','')
df['location'].fillna('N/A', inplace=True)
df['area'].fillna('N/A', inplace=True)
```

```python
# create a split string function to extract non-duplicated word 
def split_string(text):
    for i in range(1, len(text)-1):
        if text[i].isupper() and text[i-1].islower():
            return text[:i]
    return text
```

```python
df['area'] = df['area'].apply(split_string)
```

```python
df['location'] = df['location'].apply(split_string)
```

```python
df
```

```python
# standardize units for column 4
df['duration'] = df[4][df[4].str.contains(r'\d')]
```

```python
df['duration'].fillna('N/A', inplace=True)
```

```python
df['duration'] = df['duration'].apply(lambda x: x.split()[0])
```

```python
def unit_convertion (text):
    if text != 'N/A':
        unit_pattern = r"\d([a-z])"
        digit_pattern = r"\d"
        unit = re.findall(unit_pattern, text)[0]
        digit = re.findall(digit_pattern, text)[0]
        digit = int(digit)
        if unit == 'h':
            output = 0
            return output
        elif unit == 'm':
            output = digit * 30
            return output
        else:
            return digit
    else:
        return 'N/A'
        
```

```python
df['duration'] = df['duration'].apply(unit_convertion)
```

```python
# First, we extract numerical values only 
df[df[5].str.contains('\d',na=False)][5].head(50)
```

```python
# replace value containing k 
df[df[5].str.contains('\dk',na=False)][5]
```

```python
# loop through cells and multiply by 1000 if k is present
for index, salary in enumerate(df[5]):
    salary = str(salary)
    if 'k' in salary:
        pattern = r'\$?(\d+[\dk]+)'
        matches = re.findall(pattern, salary)
        for match in matches:
            if 'k' in match:
                match = match.replace('k','')
            match = float(match)
            match = match * 1000
            match = str(match)
            df.at[index,5] = match
```

```python
# replace 'to' with '-'
df[5] = df[df[5].str.contains('\d',na=False)][5].str.replace(' to ','-', regex=True)
```

```python
# This can be ignored or time 8% into the min and max salary
df[df[5].str.contains('\d\%',na=False)][5].head(100)
```

```python
# split and select from the n=1 
df[['min_salary','max_salary']] = df[5].str.split('-', n=1, expand = True).fillna('unknown')
```

```python
# extract data and clean unnessacery info
df['min_salary'] = df['min_salary'].str.extract('(\d+\.\d+|\d+ \d+|\d+k|\d+)')
df['max_salary'] = df['max_salary'].str.extract('(\d+\.\d+|\d+ \d+|\d+k|\d+)')
```

```python
df[df['min_salary'].str.contains('\d', na=False)]['min_salary'].head(100)
```

```python
df['min_salary'] = df[df['min_salary'].str.contains('\d', na=False)]['min_salary'].str.replace(' ','',regex=True)
df['max_salary'] = df[df['max_salary'].str.contains('\d', na=False)]['max_salary'].str.replace(' ','',regex=True)
```

```python
# for empty values in min and max salary
# let the unknown value equal to the known min/max value 
df.loc[df['min_salary'] == 'unknown','min_salary'] = df['max_salary']
df.loc[df['max_salary'] == 'unknown','max_salary'] = df['min_salary']
```

```python
# convert all salary into annual rate using
# first convert all data into specific data type 
df['min_salary'] = df['min_salary'].astype('float32')
df['max_salary'] = df['max_salary'].astype('float32')
```

```python
df.loc[df['min_salary'] < 50 , 'min_salary'] = df['min_salary']*8*200
df.loc[df['max_salary'] < 50 , 'max_salary'] = df['max_salary']*8*200
```
